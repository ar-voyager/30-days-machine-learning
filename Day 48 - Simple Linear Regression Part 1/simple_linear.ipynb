```python
# import matoplotlib
import matplotlib.pyplot as plt
# Import pandas
import pandas as pd
# import numpy
import numpy as np
```

```python
# Extract Data
df = pd.read_csv('placement.csv')
df.head()
```

```python
# plt.scatter()
df.plot(x='cgpa', y='package',  kind='scatter')
```

```python
# Extract 'input' and 'output' column
X = pd.DataFrame(df.iloc[:,0])
y = df.iloc[:,1]
```

```python
# from sklearn.model_selection import tarin_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20,
    random_state=2
)

# importing from sklearn.linear_models import SimpleRegression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

# train the model
lr.fit(X_train, y_train)
```

```python
# check for the test_data
print(X_test, y_test)
```

```python
# predict the value
lr.predict(X_test.iloc[[1]])
```

```python
# Orginal X_test, y_test
X_test.iloc[[1]], y_test.iloc[[1]]
```

```python
# Draw a map matplot
plt.scatter(x=df['cgpa'], y=df['package'] )
plt.plot(X_train, lr.predict(X_train), color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(In LPA)')
```

```python
# parameters of the model
# slope
m = lr.coef_
b = lr.intercept_
m, b

# Equation of the best fitted line. (y=mx+b)
y = m*8.58+b
y
```
