```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
```

```python
# Extract Data From it.
df = pd.read_csv('placement.csv')
df.head()
```

```python
# Plot the Scatter Diagram for the 'CGPA' & 'package'.
plt.scatter(df['cgpa'], df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package')
```

```python
# Extracting X, y 
X = pd.DataFrame(df.iloc[:,0])
y = df.iloc[:,1]
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.2,
    random_state = 2
)
```

```python
# Import LinearGression Model
from sklearn.linear_model import LinearRegression

# create an object of 'LinearRegression'.
lr = LinearRegression()

# Let fit the best line into the Data
lr.fit(X_train, y_train)
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.plot(X_train, lr.predict(X_train), color='red')
plt.xlabel('CGPA')
plt.ylabel('Package')
```

```python
# import metrics of linear regression from linearRegression.
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
# Mean Absolutr Percentage Error (MAPE)
from sklearn.metrics import mean_absolute_percentage_error
# Root Mean Squared Error (RMSE)
from sklearn.metrics import root_mean_squared_error

# Using the X_test data predicting target variable value.
y_pred = lr.predict(X_test)
y_pred, y_test.values
```

```python
# mean_absolute_error
# These are the metrics for which interpretation depends upon the data.
print('mean_absolute_error:',mean_absolute_error(y_test, y_pred))
print('mean_squared_error:',mean_squared_error(y_test, y_pred))
print('root_mean_squared_error:',root_mean_squared_error(y_test, y_pred))
print('mean_absolute_percentage_error:',mean_absolute_percentage_error(y_test, y_pred))

# These metric do not depend on data interpretation.
print('r2_score:',r2_score(y_test, y_pred))
```

```python
# Adjusted_R2_Square
print(X_test.shape)
# Determine r2 
r2 = r2_score(y_test, y_pred)
print('Adjusted_r2_score', 1-(1-r2)*(40-1)/(40-1-1))
```
