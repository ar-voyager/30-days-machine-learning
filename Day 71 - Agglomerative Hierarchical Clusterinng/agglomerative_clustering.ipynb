```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
# Load the Dataset
customer_data = pd.read_csv('C:\\Users\\ratha\\OneDrive\\Placement\\complete-it\\machine-learning\\30-days-machine-learning\\Day 71 - Agglomerative Hierarchical Clusterinng\\hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv') 
# Haed of the Dataset
customer_data.head()
```

```python
# Shape of the Dataset
customer_data.shape
```

```python
# Extract the Column from the Data
data = customer_data.iloc[:,3:5].values
```

```python
# Plot the Dendogram by using the scipy
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(15, 7))
plt.title('Customer Dendograms')
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
# Import Class AgglomerativeClustering
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(
    n_clusters=5,
    metric='euclidean',
    linkage='ward'
)

labels_ = cluster.fit_predict(data)

labels_
```

```python
# Plot the clusters
plt.figure(figsize=(10, 7))
plt.scatter(data[labels_ == 0,0], data[labels_ == 0,1])
plt.scatter(data[labels_ == 1,0], data[labels_ == 1,1])
plt.scatter(data[labels_ == 2,0], data[labels_ == 2,1])
plt.scatter(data[labels_ == 3,0], data[labels_ == 3,1])
plt.scatter(data[labels_ == 4,0], data[labels_ == 4,1])

# Plot the Dendogram
plt.show()
```
