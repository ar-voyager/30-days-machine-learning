```python
# import libraries
import numpy as np
import pandas as pd 
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(
    n_features=5,
    n_redundant=0,
    n_informative=5,
    n_clusters_per_class=1
)
```

```python
df = pd.DataFrame(X, columns=['col1','col2','col3','col4','col5'])
df['target'] = y 
print(df.shape)

# top rows
df.head()
```

## *Sampling Functions*

```python
# Function for row sampling
def sample_rows(df, percent): # parameters
    return df.sample(frac=percent, replace=True)
```

```python
# Function for combined sampling
def sample_features(df, percent): # Formal Arguments
    return df.sample(df.columns.tolist()[:-1], frac=percent)
```

```python
# Function for combined sampling
def combined_sampling(df, row_percent, col_percent):
    new_df = sample_rows(df, row_percent)
    return sample_features(new_df, col_percent)
```

```python
# Row Sampling
df1 = sample_rows(df, 0.2) # Actual Arguments
```

```python
df2 = sample_rows(df, 0.2) # Actual Arguments
```

```python
df3 = sample_rows(df, 0.2) # Actual Arguments: 
```

```python
df1.shape, df2.shape, df3.shape
```

```python
# import decision tree classifier
from sklearn.tree import DecisionTreeClassifier
clf1 = DecisionTreeClassifier()
clf2 = DecisionTreeClassifier()
clf3 = DecisionTreeClassifier()
```

```python
# Train the model
clf1.fit(df1.iloc[:,0:5], df1.iloc[:,-1])
clf2.fit(df2.iloc[:,0:5], df2.iloc[:,-1])
clf3.fit(df3.iloc[:,0:5], df3.iloc[:,-1])
```

```python
# plot tree 
from sklearn.tree import plot_tree
plot_tree(clf1)
```

```python
plot_tree(clf2)
```

```python
plot_tree(clf3)
```
