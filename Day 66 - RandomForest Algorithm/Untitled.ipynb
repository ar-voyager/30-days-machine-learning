```python
# Import libraries
import pandas as pd
import numpy as np

# import seaborn
import seaborn as sns
```

```python
# Dataframe
df = pd.read_csv('train.csv')
# head dataset
df.head()
```

```python
# Shape of the Dataset
df.shape
```

```python
# Extract X and y
X = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
# plot heat_map
sns.heatmap(X.iloc[5].values.reshape(28,28))
```

```python
# Get the ensemble algo RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier

# Create an Object of RandomForestClassifier
rf = RandomForestClassifier()

# Fit the model
rf.fit(X, y)
```

```python
rf.feature_importances_
```

```python
rf.feature_importances_.shape
```

```python
sns.heatmap(rf.feature_importances_.reshape(28,28))
```
