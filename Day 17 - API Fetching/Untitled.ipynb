```python
# video upload list data
rows = []

# looping
for item in range(len(data['contents'])):
     if data['contents'][item].get('type') == 'video':
         rows.append({
                'type': data['contents'][item]['type'],
                'video_id': data['contents'][item]['video'].get('videoId'),
                'title': data['contents'][item]['video'].get('title', 'NaN'),
                'length_seconds': data['contents'][item]['video'].get('lengthSeconds', 'NaN'),
                'published_time': data['contents'][item]['video'].get('publishedTimeText', 'NaN'),
                'views': data['contents'][item]['video'].get('stats', 'NaN'),
                'is_live': data['contents'][item]['video'].get('isLiveNow', 'NaN'),
                'thumbnail_url': data['contents'][item]['video'].get('thumbnails', 'NaN'),  # highest resolution
                'moving_thumbnail_url': data['contents'][item]['video'].get('movingThumbnails', 'NaN')[-1]
        })

pd.DataFrame(rows).head(2)
```
