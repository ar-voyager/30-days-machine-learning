```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# Read the data
df = pd.read_csv('titanic.csv')
```

```python
# top-5 data rows 
df.head()
```

```python
# check for the unique values of df['number']
df['number'].unique()
# value_counts
df['number'].value_counts()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passengers travelling With')
```

```python
# Extract Numerical Part
df['number_categorical'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
```

```python
# Extract Categorical Part
df['number_categorical'] = np.where(df['number_categorical'].isnull(), df['number'], np.nan)
df.head()
```

```python
df['Cabin'].unique()
df['Ticket'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\\d+)') # Capture Numerical Part
df['cabin_cat'] = df['Cabin'].str[0] # captures the first letter

# head()
df.head()
```

```python
# Now, the categories for 'cabin_cat'.
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(
    df['ticket_num'],
    errors = 'coerce',
    downcast = 'integer'
)

# extract the last bit of ticket as number
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(
    df['ticket_cat'].str.isdigit(), np.nan, df['ticket_cat'])

# head()
df.head()
```
