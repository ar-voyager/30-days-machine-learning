```python
# import libraries
import pandas as pd 
import numpy as np 
from mlxtend.plotting import plot_decision_regions
```

```python
# Create DataFrame
df = pd.DataFrame()
```

```python
# Add features to the DataFrame
df['X1'] = [1,2,3,4,5,6,6,7,9,9]
df['X2'] = [5,3,6,8,1,9,5,8,9,2]
df['label'] = [1,1,0,1,0,1,0,1,0,0]
```

```python
df
```

```python
# plot the data
import seaborn as sns
sns.scatterplot(x=df['X1'], y=df['X2'], hue=df['label'])
```

```python
# Assign Weight to the each row
df['weights'] = 1/df.shape[0]
```

```python
df
```

```python
# import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

# Create an object of DecisionTreeClassifier
# When max_depth hyperparameter is 1, tree called decision_stump
dt1 = DecisionTreeClassifier(max_depth=1)
```

```python
# Classify the data
X = df.iloc[:,0:2].values
y = df.iloc[:,2].values
```

```python
# Step 2: Train the model
dt1.fit(X, y)
```

```python
# import 
from sklearn.tree import plot_tree
plot_tree(dt1)
```

```python
# plot_decision_region
plot_decision_regions(X, y, clf=dt1, legend=2)
```

```python
# y_pred for each data_points
df['y_pred'] = dt1.predict(X)
```

```python
df
```

```python
# sum of weight for misclassified points
error = df[df.label != df.y_pred]['weights'].sum()
```

```python
def cal_model_weight(error):
    return (1/2)*np.log((1-error)/(error))
```

```python
# Step 3: Calculate model weights 
alpha1 = cal_model_weight(error)
alpha1
```

```python
# Step 4: Update Weights 
def update_row_weight(row, alpha=0.423):
    if row['label'] == row['y_pred']:
        return row['weights']*np.exp(-alpha)
    else:
        return row['weights']*np.exp(alpha)
```

```python
# callig update weigth function
df['updated_weights'] = df.apply(update_row_weight, axis=1)
```

```python

# Updated DataFrame
df
```

```python
# Normalize the weight by dividing it by sum of updated_weights
df['normalized_weight'] = df['updated_weights']/df['updated_weights'].sum()
```

```python
# Normailize weigth sum is_equal to 1
df['normalized_weight'].sum()

# Through normalize_weights, calculate lower and upper weights
# upper_weights
df['cumsum_upper'] = np.cumsum(df['normalized_weight'])
# lower_weights
df['cumsum_lower'] =  df['cumsum_upper'] - df['normalized_weight']
```

```python
# Extract specific columns
df[['X1','X2','label','weights','y_pred','updated_weights','cumsum_lower','cumsum_upper']]
```

```python
# Define a function for a random_number

def create_new_dataset(df):
    indices = [] 
    # For each row of dataset generate a random number 
    for i in range(df.shape[0]):
        a = np.random.random()
        # check random number belong which row
        for index, row in df.iterrows():
            if row['cumsum_upper'] > a and row['cumsum_lower'] < a:
                indices.append(index)
    return indices
```

```python
# generate random number 
index_values = create_new_dataset(df)
# index_values
index_values
```

```python
# new dataframe 
df2 = df.iloc[index_values, [0,1,2,3]]

# let explore
df2
```

```python
dt2 = DecisionTreeClassifier(max_depth=1)

# Split the dataset
X = df2.iloc[:,0:2].values
y = df2.iloc[:, 2].values

# Train the model
dt2.fit(X, y)
```

```python
# plot_decision_tree
plot_tree(dt2)
```

```python
# plot_decision_boundary
plot_decision_regions(X, y, clf=dt2, legend=2)
```

```python
# Get the y_pred
df2['y_pred2'] = dt2.predict(X)
```

```python
# Check for the misclassified points
df2[df2['label'] != df2['y_pred2']]
```

```python
alpha2 = cal_model_weight(0.2)
# alpha_2
alpha2
```

```python
# Step 4: Update weights 
def update_row_weights(row, alpha=5.40):
    if row['label'] == row['y_pred2']:
        return row['weights']*np.exp(-alpha2)
    else:
        return row['weights']*np.exp(alpha2)
```

```python
# Add updated_weight columns
df2['updated_weights'] = df2.apply(update_row_weights, axis=1)

# add normalized weight column
df2['normalized_weight'] = df2['updated_weights']/df2['updated_weights'].sum()

# Check for the normalize weight
df2['normalized_weight'].sum()
```
