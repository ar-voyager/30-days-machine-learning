*Adaboost Hyperparameters*

```python
# import libraries
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt

# fro scikit-learn
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_circles
```

```python
# random module
np.random.seed(42)
X, y = make_circles(
    n_samples=500, 
    factor=0.1,
    noise=0.35,
    random_state=42
)

# train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2
)
```

```python
# scatter plot
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
# import AdaboostClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import cross_val_score
```

```python
# Create Object of AdaBoostClassifier
abc = AdaBoostClassifier()

# cross_val_score
np.mean(
    cross_val_score(
        abc, X, y, scoring='accuracy', cv=10
    )
)
```

```python
# Fit the model
abc.fit(X, y)
```

```python
# define a function
def plot_decision_boundary(clf):
    plt.figure(figsize=(12, 8))
    x_range = np.linspace(X.min(), X.max(), 100)
    xx1, xx2 = np.meshgrid(x_range, x_range)
    y_hat = clf.predict(np.c_[xx1.ravel(), xx2.ravel()])
    y_hat = y_hat.reshape(xx1.shape)
    plt.contourf(xx1,xx2, y_hat, alpha=0.2)
    plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis', alpha= .7)

    # title
    plt.title('Adaboost classifier')
    plt.show()

# plot the decision_boundary
plot_decision_boundary(abc)
```

```python
# Let discuss about the hyperparameters
abc = AdaBoostClassifier(
    n_estimators = 1500, 
    learning_rate = 0.1
)

# fit the model
abc.fit(X, y)

# plot decision_boundary
plot_decision_boundary(abc)
```

```python
# GridSearchCV(Stand for CrossValidation) 
from sklearn.model_selection import GridSearchCV

grid = dict()
grid['n_estimators'] = [10, 50, 100, 500]
grid['learning_rate'] = [0.0001, 0.001, 0.1, 1.0]

# grid_search
grid_search = GridSearchCV(
    estimator=AdaBoostClassifier(), 
    param_grid = grid,
    n_jobs=1,
    cv=10,
    scoring='accuracy'
)

# execute the grid search
grid_result = grid_search.fit(X, y)

# summary the best score and configuration
print('Best: %f using %s' %(grid_result.best_score_, grid_result.best_params_))
```

```python
# plot_decision_boundary after GridSearchCV
plot_decision_boundary(grid_result)
```
