```python
class CustomRegression:
    def __init__(self):
        self.m = None
        self.b = None 
    def fit(self, X_train, y_train):
        numerator = 0 
        denominator = 0
        # iloc : search by position
        # loc: search by index_name
        for i in range(X_train.shape[0]):
            numerator += (X_train.iloc[i]-X_train.mean())*(y_train.iloc[i]-y_train.mean())
            denominator += (X_train.iloc[i]-X_train.mean())**2 

        self.m = numerator/denominator
        self.b = y_train.mean() - (self.m)*X_train.mean()
        print('Coeffi_:',self.m,'Intercept_:', self.b)

    def predict(self, X_test):
        return self.m*X_test+self.b
```

```python
# import libraries.
import numpy as np
import pandas as pd
```

```python
# Extract data from placement
df = pd.read_csv('placement.csv')
# data overview
df.head()
```

```python
# Extract column(target and independent) data from 'df'.
X = df.iloc[:,0]
y = df.iloc[:,1]
X, y
```

```python
# from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20,
    random_state = 2
)
```

## *Calculate Using the CustomRegression Class*

```python
# Predicting Using CustomRegression Class.
# Class 'CustomRegression'
lr = CustomRegression()
# Call the CustomRegression class method predict()
lr.fit(X_train, y_train)
```

```python
# Predict the data

y_pred = lr.predict(X_test.iloc[0])
y_pred
```

## *Calculate Coefficient through Sklearn*

```python
# X_train, y_train
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)

# let print the slope and intercept 
print('Coefficient_m_:',lr.coef_)
print('Intercept_b_:',lr.intercept_)
```

```python
# predict the value of X_test data and comapare with y_test.
lr.predict(X_test.iloc[[0]])
```
