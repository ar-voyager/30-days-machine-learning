```python
# Import Librraies
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X = np.random.rand(100, 1)
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100)
```

```python
# import pandas 
import pandas as pd
```

```python
df = pd.DataFrame()
df['X'] = X.reshape(100)
df['y'] = y 
```

```python
df
```

```python
# mean of the target column
df['y_pred1'] = df['y'].mean()

# Find the residual
df['res1'] = df['y'] - df['y_pred1']
```

```python
df
```

```python
# plot a scatter
plt.scatter(df['X'], df['y'])
# plot a mean line
plt.plot(df['X'], df['y_pred1'], color='red')
```

```python
# model_2
# import DecisionTreeClassifier from sklearn.tree
from sklearn.tree import DecisionTreeRegressor

tree1 = DecisionTreeRegressor(max_leaf_nodes=8)

# Fit the model
tree1.fit(df['X'].values.reshape(100, 1), df['res1'].values)
```

```python
# plot tree
from sklearn.tree import plot_tree
plot_tree(tree1)

# show the tree
plt.show()
```

```python
# generating X_test
X_test = np.linspace(-0.5, 0.5, 500)
```

```python
# Predict the values using the model_2
y_pred1 = 0.926001 + tree1.predict(X_test.reshape(500, 1))
```

```python
plt.figure(figsize=(14, 4))
plt.subplot(121)
plt.plot(X_test, y_pred1, linewidth=2, color='red')

plt.scatter(df['X'], df['y'])
```

```python
# Calculate y_pred 
df['y_pred2'] = 0.265458 + tree1.predict(df['X'].values.reshape(100,1))

# Calculate res2
df['res2'] = df['y'] - df['y_pred2']
```

**Train Model 3**

```python
# Create a object of DecisionTreeRegressor
tree2 = DecisionTreeRegressor(max_leaf_nodes=8)

# Fit the model 
tree2.fit(df['X'].values.reshape(100, 1), df['res2'].values)
```

```python
# plot_decision_tree
plot_tree(tree2)
```

```python
# y_pred 
y_pred = 0.265458 + sum(
    regressor.predict(X_test.reshape(-1, 1)) for regressor in [tree1, tree2]
)
```

```python
plt.figure(figsize=(14, 4))
plt.subplot(121)
plt.plot(X_test, y_pred, linewidth=2, color='red')
plt.scatter(df['X'], df['y'])
# plt title 
plt.title('X vs y')
```

```python
# Define a function gradient_boosting

def gradient_boost(X, y, number, lr, count=1, regs=[], foo=None):
    if number == 0:
        return
    else:
        # Do_Gradient_Boosting

        if count > 1:
            y = y - regs[-1].predict(X)
        else:
            foo = y

    tree_reg = DecisionTreeRegressor(max_leaf_nodes=5, random_state=42)
    tree_reg.fit(X, y)

    # append at the end
    regs.append(tree_reg)

    # linespace the number
    x1 = np.linspace(-0.5, 0.5, 500)

    # predict the values
    y_pred = sum(lr * regressor.predict(x1.reshape(-1, 1)) for regressor in regs)

    print(number)
    plt.figure()
    plt.plot(x1, y_pred, linwidth=2)
    plt.plot(X[:,0], foo, 'r.')
    plt.show()
    
    gradient_boost(X, y, number-1, lr, count+1, regs, foo=foo)
```

```python
np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100)
gradient_boost(X,y,5,lr=1)
```
