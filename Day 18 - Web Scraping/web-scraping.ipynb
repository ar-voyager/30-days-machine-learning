```python
# pandas and numpy library
import pandas as pd
import numpy as np
# requests library, Extracting data through API
import requests
# BeautifulSoup, for extracting Data using WebScraping
from bs4 import BeautifulSoup
```

```python
URL = 'https://www.ambitionbox.com/list-of-companies?page=1'
headers = {
    "User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'
}
webpage = requests.get(URL, headers=headers).text
```

```python
soup.find_all('h1')[0].text
```

```python
# Creating object of BeautifulSoup Class
soup = BeautifulSoup(webpage, 'lxml')
for i in soup.find_all('h2'):
    print(i.text.strip())
    
```

```python
# Extracting Company-name from Header2 tag
for i in range(1, len(soup.find_all('h2'))):
    print(soup.find_all('h2')[i].text.strip())
```

```python
# Extracting companies-rating
for rating in soup.find_all('div', class_="rating_text rating_text--md"):
    print(rating.text.strip())
```

```python
# Extracting company review 
for review in soup.find_all('span', class_="companyCardWrapper__companyRatingCount"):
    print(review.text.strip())
```

```python
# Extracting company Description
for review in soup.find_all('span', class_="companyCardWrapper__interLinking"):
    print(review.text.strip())
```

```python
# Extracting company for which Rated High
company = soup.find_all('div', class_='companyCardWrapper')
len_cmpy = len(company)
```

```python
company_name, rating, review, highly_rated, Critically_rated = [], [], [], [], []
for num in company:
    company_name.append(num.find('h2').text.strip())
    rating.append(num.find('div', class_="rating_text rating_text--md").text.strip())
    review.append(num.find('span', class_="companyCardWrapper__companyRatingCount").text.strip())
    highly_rated.append(num.find_all('span', class_="companyCardWrapper__ratingValues")[0].text.strip())
    Critically_rated.append(num.find_all('span', class_="companyCardWrapper__ratingValues")[1].text.strip())
```

```python
# Converted extracted of top-list-of-companies from AmbitionBox(Website).
# np.array(), a function of numpy python libraray.
# Ectracted lists: company_name, rating, review, highly_rated, Critically_rated
np.array([company_name, rating, review, highly_rated, Critically_rated])
```

```python
# Highly rated Companies for the reason they are known.
for i in range(len_cmpy):
    if company[i].find_all('div', class_="companyCardWrapper__ratingHeader")[0].text ==' Highly Rated For':
        highly_rated.append(company[i].find_all('span', class_="companyCardWrapper__ratingValues")[0].text)
    else:
        highly_rated.append("NA")
```

```python
# Companies Data because of that they are critised
for i in range(len_cmpy):
    if len(company[i].find_all('div', class_="companyCardWrapper__ratingHeader")) == 1:
        if company[i].find_all('div', class_="companyCardWrapper__ratingHeader")[0].text ==' Critically Rated For': 
            Critically_rated.append(company[i].find_all('span', class_="companyCardWrapper__ratingValues")[0].text)
        else:
            Critically_rated.append("NA")
    else:
        if company[i].find_all('div', class_="companyCardWrapper__ratingHeader")[1].text ==' Critically Rated For': 
            Critically_rated.append(company[i].find_all('span', class_="companyCardWrapper__ratingValues")[1].text)
        else:
            Critically_rated.append("NA")
```

```python
# Creating a DataFrame of Extracted Columns
# and Do the transpose of the DataFrame
df = pd.DataFrame([company_name, rating, review, highly_rated, Critically_rated]).transpose()
```
