```python
# import numpy & pandas
import pandas as pd
import numpy as np

# import matplotlib.pyplot
import matplotlib.pyplot as plt
# import seaborn
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
# Extracting the response and target columns
X = df.iloc[:,0:3].values
y = df.iloc[:,-1].values
```

```python
# from sklearn.model_selection import train_test_split.
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(
    X, y, test_size=0.3, random_state=1
)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# Train the model
model.fit(X_train,y_train)
```

```python
# Residual: Actual values - predicted values.
y_pred = model.predict(X_test)
residual = y_test - y_pred
```

## 1. Linear Relationship

```python
# Check for the linear.

# plot the sub-plot
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(12, 2.5))

ax1.scatter(df['feature1'], df['target'])
ax1.set_title("Feature1")
ax2.scatter(df['feature2'], df['target'])
ax2.set_title("Feature2")
ax3.scatter(df['feature3'], df['target'])
ax3.set_title("Feature3")

# show the plot
plt.show()
```

## 2. Multicollinearity

```python
# from statmodels.stats.outliers_influence
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = []

for i in range(X_train.shape[1]):
    vif.append(variance_inflation_factor(X_train, i))
```

```python
# Convert VIF into the DataFrame.
pd.DataFrame({'vif': vif}, index=df.columns[0:3]).transpose()
```

```python
# Another Technique
sns.heatmap(df.iloc[:,0:3].corr(), annot=True)
```

## 3. Normality of Residual

```python
sns.kdeplot(data=residual)
```

```python
# QQ Plot

# import scipy 
import scipy as sp

# create a plot area.
fig, ax = plt.subplots(figsize=(6,3))
sp.stats.probplot(residual, plot=ax, fit=True)

# show the plot
plt.show()
```
