```python
# Import libbraries
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import seaborn as sns

import warnings
# Ignore all warnings globally
warnings.filterwarnings("ignore")
```

```python
# Load the Dataset
df = pd.read_csv('train.csv')[['Age','Pclass','SibSp','Parch','Survived']]
```

```python
# Top-5 row of the DataFrame
df.head()
```

```python
# dropna()  --> inplace:True --> changes into the original dataframe.
df.dropna(inplace=True)
```

```python
# head()
df.head()
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
X.head()
```

```python
# cross_val_score
np.mean(cross_val_score(
    LogisticRegression(), X, y, scoring='accuracy', cv=20
))
```

## *Apply Feature Construction*

```python
X['Family_size'] = X['SibSp'] + X['Parch'] + 1
```

```python
X.head()
```

```python
def myfunc(num):
    if num == 1:
        return 0 # alone
    elif num > 1 and num <= 4:
        return 1 # small family
    else:
        return 2 # large family
```

```python
myfunc(4)
```

```python
X['Family_type'] = X['Family_size'].apply(myfunc)
X.head()
```

```python
X.drop(columns=['SibSp', 'Parch', 'Family_size'], inplace=True)
```

```python
X.head()
```

```python
# cross_val_score
np.mean(cross_val_score(
    LogisticRegression(), X, y, scoring='accuracy', cv=20
))
```
