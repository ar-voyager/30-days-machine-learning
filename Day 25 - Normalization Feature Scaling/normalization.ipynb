```python
# import libraries
import numpy as np # linear algrbra
import pandas as pd # data processing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('wine_data.csv', header=None, usecols=[0,1,2])
df.columns=['Class label', 'Alcohol', 'Malic acid']
df.sample(2)
```

```python
# kdeplot()
# Probability Density Function of 'Alcohol'
sns.kdeplot(data=df['Alcohol'])
```

```python
# Probability Density Function of 'Alcohol'
sns.kdeplot(data=df['Malic acid'])
```

```python
color_dict = {
    1:'red',
    2:'blue',
    3:'green'
}
sns.scatterplot(data=df, x=df['Alcohol'], y=df['Malic acid'], hue=df['Class label'], palette=color_dict)
```

# Featuring Scaling: Train Test Split

```python
# import sklearn.model_selection 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Class label', axis=1),
                                                     df['Class label'],
                                                     test_size=0.3,
                                                     random_state=0)
X_train.shape, X_test.shape
```

```python
# import sklearn.preprocessing
from sklearn.preprocessing import MinMaxScaler
# Create Object
scalar = MinMaxScaler()

# fit the scalar to the train set, it will learn the parameters
scalar.fit(X_train)

# transform train ans test dataset
X_train_scaled = scalar.transform(X_train)
X_test_scaled = scalar.transform(X_test)

# print data
X_train.sample(1)
```

```python
# Convert the data into the DataDrame
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
# print data
X_train_scaled.sample(2)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# scatter plot ax1
ax1.scatter(X_train['Alcohol'], X_train['Malic acid'], c=y_train)
ax1.set_title('Before Scaling')

# scatter plot ax2
ax2.scatter(X_train_scaled['Alcohol'], X_train_scaled['Malic acid'], c=y_train)
ax2.set_title('After Scaling')

# show graph
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# Before Scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Alcohol'], ax=ax1)
sns.kdeplot(X_train['Malic acid'], ax=ax1)

# After Scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_scaled['Alcohol'], ax=ax2)
sns.kdeplot(X_train_scaled['Malic acid'], ax=ax2)

# plot
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# Before Scaling
ax1.set_title('Alcohol Distribution Before Scaling')
sns.kdeplot(X_train['Alcohol'], ax=ax1)

# After Scaling
ax2.set_title('Alcohol Distribution After Standard Scaling')
sns.kdeplot(X_train_scaled['Alcohol'], ax=ax2)

# plot
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# Before Scaling
ax1.set_title('Malic acid Dist Before Scaling')
sns.kdeplot(X_train['Malic acid'], ax=ax1)

# After Scaling
ax2.set_title('Malic acid Dist After Standard Scaling')
sns.kdeplot(X_train_scaled['Malic acid'], ax=ax2)

# plot
plt.show()
```
